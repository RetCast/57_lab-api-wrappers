### 8.05 Activity 2

- Create a function to extract the id of a song
- Create a function to extract the audio features of a song
- Create a dataframe with the information, including the name of the song and the artist

```python
list_of_songs = ['Paper Planes', 'Die Hard', 'Fashion Killa', 'Topdown', 'Smack That', 'Stronger', 'Take Care', 'DNA', "Wouldn't Get Far", "Get Like Me"]
list_of_songs
```

#### Getting songs ids

```python
def songs_ids(list_of_songs):
    songs_ids = []
    for song in list_of_songs:
        song_result = sp.search(q=song, limit=1)
        track_id = song_result['tracks']['items'][0]['id']
        songs_ids.append(track_id)
    return songs_ids
```

```python
songs_ids(list_of_songs)
```

#### Getting song features

```python
song = sp.search(q=list_of_songs[0], limit=1)
song
```

```python
song['tracks']['items'][0]['artists'][0]['name']
```

```python
audio_f = sp.audio_features(song['tracks']['items'][0]['id'])[0]
audio_f
```

```python
audio_df = pd.DataFrame(audio_f, index=[0])
audio_df
```

```python
audio_df['track'] = song['tracks']['items'][0]['name']
audio_df['artist'] = song['tracks']['items'][0]['artists'][0]['name']
audio_df
```

```python
def audio_features(list_of_songs):
    audio_f = []
    
    for song in list_of_songs:
        song_result = sp.search(q=song, limit=1)
        track_id = song_result['tracks']['items'][0]['id']
        
        audio_info = sp.audio_features(track_id)[0]
        audio_info['track'] = song_result['tracks']['items'][0]['name']
        audio_info['artist'] = song_result['tracks']['items'][0]['artists'][0]['name']
        
        audio_f.append(audio_info)
    
    audio_df = pd.DataFrame(audio_f)
    return audio_df
```

```python
audio_features(list_of_songs)
```

### Song Recommender version 1

```python
list_songs = pd.read_csv('top_100_billboard_songs_2023-08-12.csv')
list_songs.head(40)
```

```python
def search_song():
    while True:
        search_song = input("Hello! Please tell me a song you like: ")
        search_song = search_song.lower()
        print()

        # Search the song in the DataFrame
        matching_songs = list_songs[list_songs['song'].apply(lambda x: x.lower()).str.contains(search_song)]

        if not matching_songs.empty:
            print("We found the following matching songs:")
            for i, song_row in enumerate(matching_songs.iterrows(), start=1):
                song_name = song_row[1]['song']
                artist_name = song_row[1]['artist']
                print(f"{i}. {song_name} by {artist_name}")
            
            user_choice = input("Please enter the number of the song you meant, or 'no' if none of these are correct: ")

            if user_choice.lower() == 'no':
                print()
                print("Sorry about that. Let's try again.")
                print()
            elif user_choice.isdigit() and 1 <= int(user_choice) <= len(matching_songs):
                chosen_song = matching_songs.iloc[int(user_choice) - 1]
                artist_songs = list_songs[list_songs['artist'] == chosen_song['artist']]
                if not artist_songs.empty:
                    artist_songs = artist_songs[artist_songs['song'] != chosen_song['song']]  # Exclude the chosen song
                    if not artist_songs.empty:
                        random_recommendation = artist_songs.sample(n=1).iloc[0]
                        print()
                        print("Here's a song by the same artist that you might like:")
                        print(f"Song: {random_recommendation['song']}")
                        print(f"Artist: {random_recommendation['artist']}")
                    
                break
            else:
                print("Invalid input. Please enter a valid number or 'no'.")
        else:
            print("No matching songs found. Please try again.")
            print()
```
